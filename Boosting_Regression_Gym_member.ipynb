# Get Dummies

```python
obj=pd.get_dummies(obj,columns=["Workout_Type"]).astype(int)
```

```python
obj
```

```python
num
```

```python
from sklearn.preprocessing import MinMaxScaler
q=MinMaxScaler()
for i in num:
    num[i]=q.fit_transform(num[[i]])
```

```python
num.corr().tail(1)
```

```python
new_df=pd.concat([obj,num],axis=1)
new_df
```

```python
x=new_df.drop("Calories_Burned",axis=1)
y=new_df["Calories_Burned"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=11)
```

# AdaBoostRegressor

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
adb=AdaBoostRegressor(random_state=11)
```

```python
adb.fit(x_train,y_train)
```

```python
y_pred=adb.predict(x_test)
```

```python
adb.score(x_test,y_test)
```

```python
adb.score(x_train,y_train)
```

# GradientBoostingRegressor
